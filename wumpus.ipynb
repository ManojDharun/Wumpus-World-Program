```python
def printpath(wumpus,path):
    for i in range(0,len(wumpus)):
        for j in range(0,len(wumpus[i])):
            if(i==path[0] and path[1] == j):
                print("A",end="      ")
            elif(wumpus[i][j].stench == 1 and wumpus[i][j].breeze == 1 and wumpus[i][j].gold == 1):
                print("St/Br/G",end =" ")
            elif(wumpus[i][j].stench == 1 and wumpus[i][j].breeze == 1):
                print("St/Br",end ="  ")
            elif(wumpus[i][j].stench==1):
                print("St",end = "     ")
            elif(wumpus[i][j].breeze == 1):
                print("Br",end= "     ")
            elif(wumpus[i][j].wumpus == 1):
                print("!W!",end="    ")
            elif(wumpus[i][j].pit==1):
                print("*P*",end="    ")
            elif(wumpus[i][j].gold == 1):
                print("Gold",end="   ")
            else:
                print("-",end="       ")
        print()
    print()
    print("Agent's current position ",path)
    print()
```
